# 🔹 2. Parallelism in Python

Parallelism is useful for **CPU-bound tasks**, like:

* Image processing
* Machine learning
* Heavy calculations

Python uses **multiprocessing** for true parallelism because of the **GIL (Global Interpreter Lock)**.

```python
from multiprocessing import Pool
import time

def square(n):
    time.sleep(1)
    return n * n

if __name__ == "__main__":
    with Pool(4) as p:
        result = p.map(square, [1, 2, 3, 4])
    print(result)
```

### 🔍 What happens:

* 4 processes run at the same time
* Each task takes 1 sec
* Total time ≈ 1 sec (instead of 4 sec)

---

# 🔹 Important Note (GIL)

Python has something called the **Global Interpreter Lock (GIL)**:

* It allows only **one thread to execute Python bytecode at a time**
* So:

  * Threads ≠ true parallelism for CPU tasks
  * Use **multiprocessing** for real parallel execution

---

# 🔹 When to use what?

* Use **asyncio / threading** 👉 when waiting (network, disk)
* Use **multiprocessing** 👉 when computing (CPU heavy)
