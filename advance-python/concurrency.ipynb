* **Concurrency** → dealing with *multiple tasks at once* (but not necessarily at the same exact time)
* **Parallelism** → *actually running tasks simultaneously* (at the same time, using multiple CPUs)

# 🔹 1. Concurrency in Python

Concurrency is useful when tasks spend time waiting (I/O-bound tasks), like:

* API calls
* File reading/writing
* Network requests

```python
import asyncio


async def task(name, delay):
    print(f"Start {name}")
    await asyncio.sleep(delay)
    print(f"End {name}")


async def main():
    await asyncio.gather(
        task("Task 1", 2),
        task("Task 2", 1),
        task("Task 3", 3)
    )

asyncio.run(main())
```

### 🔍 What happens:

* All tasks start together
* They don’t block each other while waiting
* Total time ≈ longest task (3 sec), not sum (6 sec)

# 🔹 Real-world use case (Concurrency)

👉 **Web scraping / API calls**

Example:

* Fetching data from 100 websites
* Instead of waiting one-by-one, you send all requests together

Used in:

* Chat apps (handling many users)
* Web servers
* Bots
